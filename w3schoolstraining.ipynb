### variabile globale

```python
print("Variabilele globale sunt create în afara unei funcții și pot fi utilizate oriunde în program, inclusiv în interiorul funcțiilor. Variabilele locale sunt create în interiorul unei funcții și pot fi utilizate doar în acea funcție.")
```

```python
x="merge"
def myfunc():
    global x
    print("Python is " + x)
myfunc()
```

```python
y="creez"
def functiamea():
    global y
    print("Astazi " + y + " o noua functie cu variabila globala")
functiamea()
```

```python
x = "bun"

def myfunc():
  x = "superb"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

```python
def functia():
    global x
    x = "superb"
functia()
print("Python is " + x)
```

```python
x="Variabila globala normala inainte de functie"
def func():
    global x
    x="Variabila globala modificata in functie"
func()
print(x)
```
